# scitex-genai — LLM Quick Start

Minimal walk-through of the unified `GenAI` factory over OpenAI / Anthropic / Google / Groq / DeepSeek / Perplexity / Llama. Same call shape for every provider — pass any model name from the catalogue, the factory dispatches to the matching provider class.

**What this notebook covers**

1. List available models.
2. Initialise a `GenAI` client (no API call yet).
3. Run a single completion via `__call__`.
4. Inspect cost and history.

Each cell skips gracefully if the relevant API key (e.g. `OPENAI_API_KEY`) is unset.

```python
import os
import scitex_genai

scitex_genai.__version__
```

## 1. Browse the model catalogue

`MODELS` is a DataFrame of every supported `(name, provider)` pair. The factory looks up the provider from the model name.

```python
from scitex_genai.llm._PARAMS import MODELS
MODELS[['name', 'provider']].head(10)
```

## 2. Initialise a provider client

Pass any catalogue model name. `GenAI` dispatches to the right provider class (`OpenAI`, `Anthropic`, ...) under the hood.

```python
if not os.environ.get('OPENAI_API_KEY'):
    print('OPENAI_API_KEY not set — using stub. Set the key to run a real call.')
    ai = None
else:
    ai = scitex_genai.GenAI(model='gpt-4o-mini')
    print('client ready:', type(ai).__name__, '·', ai.model)
```

## 3. Single completion

The instance is callable (`ai(prompt)`); the response is returned as text.

```python
if ai is not None:
    response = ai('Summarise neural networks in one sentence.')
    print(response)
else:
    print('skipped — no API key.')
```

## 4. Cost + history

`ai.cost` is a USD estimate accumulated across calls. `ai.history` is the running conversation buffer; reset it with `ai.reset()`.

```python
if ai is not None:
    print('cost (USD):', ai.cost)
    print('history turns:', len(ai.history))
else:
    print('skipped — no API key.')
```

`scitex_genai` is organised by modality so the namespace ages well. Today only `llm` is implemented — these stubs are importable but raise `NotImplementedError`:

- `scitex_genai.agent` — agentic loops (claude-agent-sdk wrapper planned)
- `scitex_genai.image` — image generation / editing
- `scitex_genai.audio` — TTS / STT / music
- `scitex_genai.video` — video generation
- `scitex_genai.embed` — embeddings
- `scitex_genai.multimodal` — any-to-any unified models
