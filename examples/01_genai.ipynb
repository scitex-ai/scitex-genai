# scitex-genai — LLM Quick Start

Minimal walk-through of the unified `GenAI` factory over OpenAI / Anthropic / Google / Groq / DeepSeek / Perplexity / Llama. Same API for every provider, with cost tracking and conversation history.

**What this notebook covers**

1. Initialise a `GenAI` client.
2. Run a single completion.
3. Inspect the cost summary.

Each cell skips gracefully if the relevant API key (e.g. `OPENAI_API_KEY`) is not set.

```python
import os
import scitex_genai

scitex_genai.__version__
```

## 1. Initialise a provider client

```python
if not os.environ.get('OPENAI_API_KEY'):
    print('OPENAI_API_KEY not set — using stub. Set the key to run a real call.')
    ai = None
else:
    ai = scitex_genai.GenAI(provider='openai', model='gpt-4o-mini')
    print('client ready:', ai)
```

## 2. Single completion

```python
if ai is not None:
    response = ai.complete('Summarise neural networks in one sentence.')
    print(response)
else:
    print('skipped — no API key.')
```

## 3. Cost summary

```python
if ai is not None:
    print(ai.get_cost_summary())
else:
    print('skipped — no API key.')
```
